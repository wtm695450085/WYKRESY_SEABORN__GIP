```python
def Dyst6(x, ax, title, x_label, y_label):
    x.plot.kde(ax=ax, legend=True)
    x.plot.hist(density=True, ax=ax)
    ax.set_title(title, color="darkred", alpha=1)
    ax.set_ylabel(y_label, color="grey", alpha=0.6)
    ax.set_xlabel(x_label, color="grey", alpha=0.6)
    ax.grid(color="grey", alpha=0.3)
```

```python
def Dyst7(x, y, ax, title, x_label, y_label):
    x.plot.kde(ax=ax, legend=True)
    x.plot.hist(density=True, ax=ax)
    y.plot.kde(ax=ax, legend=True)
    ax.set_title(title, color="darkred", alpha=1)
    ax.set_ylabel(y_label, color="grey", alpha=0.6)
    ax.set_xlabel(x_label, color="grey", alpha=0.6)
    ax.grid(color="grey", alpha=0.3)
```

```python
def Dist1(x, ax, title, x_label, y_label):
    x.plot.kde(ax=ax, legend=True)
    ax.set_title(title, color="darkred", alpha=1)
    ax.set_ylabel(y_label, color="grey", alpha=0.6)
    ax.set_xlabel(x_label, color="grey", alpha=0.6)
    ax.grid(axis="y", color="grey", alpha=0.6)
```

```python
def Dyst8(x, y, ax, title, x_label, y_label):
    x.plot.kde(ax=ax, legend=True)
    x.plot.hist(density=True, ax=ax, color="y", alpha=0.3)
    y.plot.kde(ax=ax, legend=True)
    y.plot.hist(density=True, ax=ax, color="b", alpha=0.3)
    ax.set_title(title, color="darkred", alpha=1)
    ax.set_ylabel(y_label, color="grey", alpha=0.6)
    ax.set_xlabel(x_label, color="grey", alpha=0.6)
    ax.grid(color="grey", alpha=0.3)
```

```python
def WP2(kat1, kat2, kat3, x, y, ax, title, x_label, y_label):
    kat1.plot(x=x, y=y, ax=ax, label="Poland", color="r")
    kat2.plot(x=x, y=y, ax=ax, label="Germany", color="b")
    kat3.plot(x=x, y=y, ax=ax, label="France", color="g")
    ax.set_title(title, color="darkred", alpha=1)
    ax.set_ylabel(y_label, color="grey", alpha=0.6)
    ax.set_xlabel(x_label, color="grey", alpha=0.6)
    ax.grid(color="grey", alpha=0.3)
    ax.set_facecolor("#d8dcd6")
```

```python
kat1 = df[df["Country"] == "Poland"]
kat2 = df[df["Country"] == "Germany"]
kat3 = df[df["Country"] == "France"]

x = "Year"
y = "Happiness Score"
title = "Poland - Happiness Score"
x_label = "year"
y_label = "level"
```

```python
fig, ax = plt.subplots(figsize=(8, 4))
WP2(kat1, kat2, kat3, x, y, ax, title, x_label, y_label)
```

```python
def WWP1(kat1, kat2, x, y):
    fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, figsize=(8, 4))

    kat1.plot(x=x, y=y, legend=False, ax=ax1)
    ax1.set_title("Poland")

    kat2.plot(x=x, y=y, legend=False, ax=ax2)
    ax2.set_title("France")

    plt.tight_layout()
```
