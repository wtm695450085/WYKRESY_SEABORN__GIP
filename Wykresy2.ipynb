```python
fig, (ax1, ax2, ax3, ax4) = plt.subplots(nrows=1, ncols=4, figsize=(16, 4))

grid = plt.GridSpec(2, 3, wspace=0.2, hspace=0.3)
ax1 = plt.subplot(grid[0, 0])
ax2 = plt.subplot(grid[0, 1:])
ax3 = plt.subplot(grid[1, :1])
ax4 = plt.subplot(grid[1, 1:])

ax1.plot(df["Freedom"], color="green", alpha=0.8)
ax1.set_title("Freedom in the world", color="darkblue")
ax1.set_xlabel("jakaś zmienna")
ax1.set_ylabel("ogólnie wykres bez sensu", color="brown")

ax2.plot(df["Generosity"], color="black", alpha=0.8)
ax2.plot(df["Economy (GDP per Capita)"], color="blue", alpha=0.8)
ax2.legend(loc=(0.65, 0.8))
ax3.scatter(df["Generosity"], df["Freedom"], color="blue", alpha=0.8)
ax3.legend(loc=(0.65, 0.8))
ax4.plot(df["Family"], color="brown", alpha=0.8)
ax4.plot(df["Health (Life Expectancy)"], color="black", alpha=0.8)
```
